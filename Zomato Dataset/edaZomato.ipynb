```python
df_country = pd.read_excel("/workspaces/codespaces-jupyter/data/Zomatodataset/Country-Code.xlsx")
df_country.head()
```

```python
# Merging the two datasets

final_df = pd.merge(df, df_country, on='Country Code', how='left')
final_df.head()
```

```python
# Check datatypes
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
# Pie Chart - Top 3 countries that use Zomato
plt.pie(country_values[:3], labels=country_names[:3], autopct= '%1.2f%%')
```

```python
# Increase the size of the plot
plt.figure(figsize=(12,6))
```

### **Observation:** 

1. Zomato maximum records or transaction are from India
2. Second comes The United States of America
3. Third is the United Kingdom


```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

### **Observation:**

1. Rating between 4.5 and 4.9 -> Excellent
2. Rating between 4.0 and 4.4 -> Very Good
3. Rating between 3.5 and 3.9 -> Good
4. Rating between 4.0 and 3.4 -> Average
5. Rating between 2.5 and 2.9 -> Average
6. Rating below 2 -> Poor
7. Rating 0 -> Not rated

```python
ratings.head()
```

```python
[color for color in ratings['Rating color'].unique()]
```

```python
# Plot the ratings count with respect to the aggregate rating with the approperiate color
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings, hue='Rating color', palette=['Blue', 'Red', 'Orange', 'Yellow', 'Green', 'Green'])

# Increase the size of the sns barplot
plt.figure(figsize=(12,6))
```

---
### **Observation:**

1. Not rated count is very high
2. Maximum number of ratings are between 2.5 to 3.4
---

```python
## Count plot

sns.countplot(x='Rating color', data=ratings, palette = ['Blue', 'Red', 'Orange', 'Yellow', 'Green', 'Green'])
```

#### **Top 10 Cuisines**

```python
# Top 10 Cuisines in the world
final_df.groupby('Cuisines').size().reset_index().rename(columns={0:'Count'}).sort_values(by='Count', ascending=False)[:10]
```
